#### A very simple example from http://stackoverflow.com/q/19073683/1304161, but I'll show some special tricks with it
First let's just create the data and plot it with no adjustment or anything.

```python
together = [(0, 1.0, 0.4), (25, 1.0127692669427917, 0.41), (50, 1.016404709797609, 0.41), (75, 1.1043426359673716, 0.42), (100, 1.1610446924342996, 0.44), (125, 1.1685687930691457, 0.43), (150, 1.3486407784550272, 0.45), (250, 1.4013999168008104, 0.45)]
together.sort()

text = [x for (x,y,z) in together]
eucs = [y for (x,y,z) in together]
covers = [z for (x,y,z) in together]
def plot_eucs_covers():
    plt.plot(eucs,covers,color="black", alpha=0.5)
    texts = []
    for xt, yt, s in zip(eucs, covers, text):
        texts.append(plt.text(xt, yt, s))

    plt.xlabel("Proportional Euclidean Distance")
    plt.ylabel("Percentage Timewindows Attended")
    plt.title("Test plot")
    return texts
plot_eucs_covers()
plt.show()
```

Now let's just apply the text adjustment with all default parameters and define the arrow properties we like.

```python
texts = plot_eucs_covers()
adjust_text(texts, arrowprops=dict(arrowstyle="->", color='r', lw=0.5))
plt.show()
```

Now one of the labels overlays the line, that's not pretty. We'll create some virtual points along the lines using linear interpolation to repel the labels from the graph itself.

```python
from scipy import interpolate
texts = plot_eucs_covers()
f = interpolate.interp1d(eucs, covers)
x = np.linspace(min(eucs), max(eucs), 1000)
y = f(x)
adjust_text(texts, x, y, arrowprops=dict(arrowstyle="->", color='r', lw=0.5))
plt.show()
```

Well, we don't really want the labels to move along the x axis that far! Let's prevent any movement along the x axis completely.

```python
texts = plot_eucs_covers()
adjust_text(texts, x, y, arrowprops=dict(arrowstyle="->", color='r', lw=0.5),
            autoalign='y', only_move={'points':'y', 'text':'y'})
plt.show()
```

That's kind of OK, but two texts overlap! They have the same y coordinate, and we forbade any movement along the x axis, so this is a bit tricky, because they don't repel each other in y... And apparently autoalignement places them on the same side of the line due to our virtual points. Let's play around with the parameters, maybe those virtual points can help us somehow, because these labels will interact with them a bit differently. To tune that the most important is to get them to be on different sides of the line during autoalignment, no way they can cross the line during the adjustment stage. For that we need to find an optimal expand_align value, so that the labels intersect the virtual points very differently. I manually found the best value is (0.2, 0.2).

We also want the labels to end up a little further away from the lines, but at the same time not fly so far. For that we need to expand text bounding box when repelling from points more (in our case we need to make it higher), but reduce the force of repelling from points - then the labels should end up further away from the points, but will move slower and won't fly away very far.

```python
texts = plot_eucs_covers()
adjust_text(texts, x, y, arrowprops=dict(arrowstyle="->", color='r', lw=0.5),
            autoalign='y', only_move={'points':'y', 'text':'y'}, expand_align=(0.2, 0.2), expand_points=(1.2, 1.5),
            force_points=0.15)
plt.show()
```

Now this is essentially perfect! Dealing with unequal distance of labels to the line is difficult because of the artificial way we are repelling them and different shape of the line close to different labels.

###### Now a real-world example of a volcano plot
(idea and data from http://www.gettinggeneticsdone.com/2016/01/repel-overlapping-text-labels-in-ggplot2.html)

First let's just plot the data 

```python
data = pd.read_csv('volcano_data.csv')
def plot_volcano():
    plt.figure(figsize=(7, 10))
    threshold = 0.05
    xns, yns = data['log2FoldChange'][data['padj']>=threshold], -np.log10(data['pvalue'][data['padj']>=threshold])
    plt.scatter(xns, yns, c='grey', edgecolor=(1,1,1,0), label='Not Sig')
    xs, ys = data['log2FoldChange'][data['padj']<threshold], -np.log10(data['pvalue'][data['padj']<threshold])
    plt.scatter(xs, ys, c='r', edgecolor=(1,1,1,0), label='FDR<5%')
    texts = []
    for x, y, l in zip(xs, ys, data['Gene'][data['padj']<threshold]):
        texts.append(plt.text(x, y, l, size=8))
    plt.legend()
    plt.xlabel('$log_2(Fold Change)$')
    plt.ylabel('$-log_{10}(pvalue)$')
    return texts
plot_volcano()
plt.show()
```

We can't read half of the labels in the left half! Let's adjust them :)

```python
texts = plot_volcano()
adjust_text(texts, arrowprops=dict(arrowstyle="-", color='k', lw=0.5))
plt.show()
```

What else could you want from this? It's great! And we got that by just adding one line of code. If you don't like that some labels ended up a bit far away from the points, we can change that by reducing either force_text, or expand_text (especially in x in this case), or both:

```python
texts = plot_volcano()
adjust_text(texts, force_text=0.5, expand_text=(1., 1.2), arrowprops=dict(arrowstyle="-", color='k', lw=0.5))
plt.show()
```

#### Now a classical mtcars dataset; example from ggrepel package for R

```python
mtcars = pd.read_csv('mtcars.csv')
def plot_mtcars():
    plt.figure(figsize=(9, 6))
    plt.scatter(mtcars['wt'], mtcars['mpg'], s=15, c='r', edgecolors=(1,1,1,0))
    texts = []
    for x, y, s in zip(mtcars['wt'], mtcars['mpg'], mtcars['Car']):
        texts.append(plt.text(x, y, s, size=7))
    return texts
plot_mtcars()
plt.show()
```

Aaa everything is overlapping!!! Crazy. Let's fix it.

```python
texts = plot_mtcars()
adjust_text(texts, arrowprops=dict(arrowstyle="-", color='k', lw=0.5))
plt.show()
```

Well... Default settings clearly don't work that well here, it looks a little messy and some labels are unnecessarily far away. What usually helps in this case is reducing the repelling force - the labels then move slower and get a chance to adjust better. Let's try with force of repelling from points. Also everything is pretty dense here, so let's expand the labels in x direction a bit less (default is 1.2 times in all situations). At the same time we want the autoalignement to be sensitive to labels and points a bit further away than normally, so let's increase expand_align.

```python
texts = plot_mtcars()
adjust_text(texts, arrowprops=dict(arrowstyle="-", color='k', lw=0.5),
            expand_points=(1.1, 1.2), expand_text=(1.1, 1.2),
            expand_align=(1.5, 1.5), force_points=0.5, force_text=0.5)
plt.show()
```

Now this is pretty good for such a difficult case!

Note that I specified the figure size for two latter examples. Obviously, making it larger is one of the easiest ways to reduce overlaps, but often it is limited - that's when this library can help!

#### Now an example from https://blog.mafr.de/2012/03/11/time-series-data-with-matplotlib/ with adding labels

```python
import matplotlib.dates as mdates
import locale
locale.setlocale(locale.LC_ALL,'en_GB.utf8') #I have it set to Russian, without this dates don't work
```

```python
days, impressions = np.loadtxt("page-impressions.csv", unpack=True,
        converters={ 0: mdates.strpdate2num('%Y-%m-%d')})
```

```python
def plot_dates(adjust=False):
    plt.figure(figsize=(10, 7))
    plt.plot_date(x=days, y=impressions)
    texts = []
    for x, y in zip(days, impressions):
        texts.append(plt.text(x, y, y))
    if adjust:
        adjust_text(texts, arrowprops=dict(arrowstyle="->", color='r', lw=0.5))
    plt.show()
plot_dates(adjust=False)
```

```python
plot_dates(adjust=True) #As a bonus forces labels inside the axes
```

Practically perfect! The one arrow that is a bit longer than others can be fixed with lower force_text (not showing it here, you can experiment yourself).
